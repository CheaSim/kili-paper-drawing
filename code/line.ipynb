```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager as fm
myfont = fm.FontProperties(fname='simsun.ttc')
```

```python
data = pd.read_excel("./result.xlsx")
```

```python

#params
x = "xxx"
y = "yyy"
title = "result"



#end params


plt.figure(figsize=(6, 5.2))
sns.set(style='whitegrid', color_codes=True)
sns.set_context("paper")
plt.rcParams.update({'font.size': 6})
# sns.color_palette("bright",10)
plt.rcParams['font.family'] = ['Times New Roman'] 
plt.rcParams.update({'legend.loc': "lower right"}) 
plt.rcParams.update({'legend.fontsize': 13.5}) 
# plt.rcParams.update({'legend.title': None}) 

ax1 = sns.lineplot(x="x", y="y", hue="model", data=data, lw=4,
   marker="*",
    sort=False,
)

plt.title(title, fontsize=20)
def set_sc():
    # show the sc
    ax1.spines['top'].set_visible(True)
    ax1.spines['right'].set_visible(True)
    ax1.spines['bottom'].set_visible(True)
    ax1.spines['left'].set_visible(True)
    # set the style of sc
    ax1.spines['top'].set_linewidth('2.0')#设置边框线宽为2.0
    ax1.spines['right'].set_linewidth('2.0')#设置边框线宽为2.0
    ax1.spines['bottom'].set_linewidth('2.0')#设置边框线宽为2.0
    ax1.spines['left'].set_linewidth('2.0')#设置边框线宽为2.0
    
    ax1.spines['top'].set_color('black')
    ax1.spines['right'].set_color('black')
    ax1.spines['bottom'].set_color('black')
    ax1.spines['left'].set_color('black')
    
    ax1.xaxis.label.set_size(16)
    ax1.yaxis.label.set_size(16)
    ax1.xaxis.label.set_fontname('Times New Roman')
    ax1.xaxis.label.set_fontname('Times New Roman')
    plt.tick_params(labelsize=18)
    labels = ax1.get_xticklabels() + ax1.get_yticklabels() 
    [label.set_fontname('Times New Roman') for label in labels]


# plt.rcParams.update({'legend.title': None}) 
set_sc()

# plt.grid(ls='--',linewidth=3)
d = {"8":8, "16":16,"32":32, "full":60}

ax1.set_xticks(list(d.values()))
ax1.set_xticklabels(d.keys())
plt.xlabel(x, fontsize=18)
plt.ylabel(y, fontsize=18)
ax1.tick_params(
    direction='out', # 刻度线朝内
    length=6, width=2, # 长度和宽度
    bottom=True,
    left=True,
)

# 右下角 图例介绍设置
handles, labels = ax1.get_legend_handles_labels()
rule = {0:3,1:0,2:1,3:2}
handles, labels = handles[1:], labels[1:]
handles = sorted(handles, key=lambda x:rule[handles.index(x)])
labels = sorted(labels, key=lambda x:rule[labels.index(x)])
labels = ["a","b","c","d"]
x = plt.legend(handles, labels)
# plt.legend()
leg = x
leg.get_lines()[0].set_linewidth(3)
leg.get_lines()[1].set_linewidth(3)
leg.get_lines()[2].set_linewidth(3)
leg.get_lines()[3].set_linewidth(3)
# plt.legend(*sc.legend_elements(), bbox_to_anchor=(1.05, 1), loc=2)
# plt.legend(*my_legend, bbox_to_anchor=(1.05, 1), loc=0)
plt.savefig("result.pdf")
plt.show()
```
